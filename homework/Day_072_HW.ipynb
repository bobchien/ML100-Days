# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

def ReLU(x):
    return list(map(lambda x:max(0, x), x))
def dReLU(x):
    return list(map(lambda x:0 if x<=0 else 1, x))


x = plt.linspace(-10, 10, 100)
plt.plot(x, ReLU(x), 'b', label='linspace(-10, 10, 100)', color='blue')
plt.plot(x, dReLU(x), 'b', label='linspace(-10, 10, 100)', color='yellow', alpha=0.6)
plt.grid()
plt.title('ReLU Function')
plt.legend(['ReLU', 'dReLU'])
plt.text(-9.5, 5, r'ReLU(x)=Max(0, x)', fontsize=15)

plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(1))
plt.show()
```
