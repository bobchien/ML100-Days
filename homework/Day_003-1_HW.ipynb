### Hints: [隨機產生數值](https://blog.csdn.net/christianashannon/article/details/78867204)

```python
import pandas as pd
import numpy as np
```

```python
basic_people = 10000000
countries = []
for a in "ABCDEFGHIJKLMNOPQRSTUVWXYZ":
    countries.append(a)
people = (np.random.rand(len(countries)) * basic_people).astype(int)
```

```python
data = {'國家': countries,
        '人口': people}
data = pd.DataFrame(data)
data
```

```python
data.max()
```
