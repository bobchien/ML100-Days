使用function y=(x+5)² 來練習學習率( leanrning rate ) 調整所造成的影響

# 作業目標:
    請嘗試使用不同的組合驗證 learning rate 對所需 iteration 的影響
    lr = [0.1, 0.0001]
    主要驗證 Lr 對於grandient 收斂的速度
    

```python
import matplotlib.pyplot as plt

precision = 0.000001 #This tells us when to stop the algorithm
max_iters = 10000 # maximum number of iterations
df = lambda x: 2*(x+5) #Gradient of our function 
```

```python
lr = [0.5, 0.1, 0.01, 0.001, 0.0001]
colors = ['yellow', 'red', 'blue', 'green', 'black']
for i in range(len(lr)):
    old_x = 3
    iters = 0
    previous_step_size = 1
    x_history = [old_x]
    iter_history = [iters]
    while previous_step_size > precision and iters < max_iters:
        new_x = old_x - lr[i] * df(old_x)
        previous_step_size = abs(old_x - new_x)
        iters +=1
        old_x = new_x
        x_history.append(old_x)
        iter_history.append(iters)
        #print("Iteration", iters,": X value is", old_x)
    plt.plot(iter_history, x_history, 'o-', ms=3, lw=1.5, color=colors[i])
    plt.xlabel(r'$iters$', fontsize=16)
    plt.ylabel(r'$x$', fontsize=16)
    plt.show()
    print("The learning rate is", lr[i], "\nTotel iteration number is", iters, "\nThe value of X value is", old_x)
```
