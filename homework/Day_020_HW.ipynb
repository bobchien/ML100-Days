# 作業 : (Kaggle)房價預測

```python
# 做完特徵工程前的所有準備 (與前範例相同)
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

data_path = 'data/'
df_train = pd.read_csv(data_path + 'house_train.csv.gz')

train_Y = np.log1p(df_train['SalePrice'])
train_Y_o = train_Y.copy()
df = df_train.drop(['Id', 'SalePrice'] , axis=1)
df.head()
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features : {num_features}\n')
```

```python
# 削減文字型欄位, 只剩數值型欄位
df = df[num_features]
df = df.fillna(-1)
MMEncoder = MinMaxScaler()
train_num = train_Y.shape[0]
df_o = df.copy()
df.head()
```

# 作業1
* 試著限制 '1樓地板面積(平方英尺)' (1stFlrSF) 欄位的上下限, 看看能否再進一步提高分數?
* 答案：可以，透過第10百分位數及第95百分位數，可將分數從0.8434973274955462提升至0.8531096740599757。

```python
# 顯示 1stFlrSF 與目標值的散佈圖
import seaborn as sns
import matplotlib.pyplot as plt
sns.regplot(x = df['1stFlrSF'][:train_num], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

```python
df['1stFlrSF'][:train_num].describe()
```

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 調整離群值
df = df_o.copy()
df['1stFlrSF'][:train_num] = df['1stFlrSF'][:train_num].clip(df['1stFlrSF'][:train_num].quantile(q=0.10), df['1stFlrSF'][:train_num].quantile(q=0.90))

sns.regplot(x = df['1stFlrSF'][:train_num], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

* 續前題, 去除離群值有兩類方式 :  捨棄離群值(刪除離群的資料) 以及調整離群值,

* 答案：透過第10百分位數及第95百分位數作為上、下限，調整離群值後，可將分數從0.8531096740599757更進一步提升至0.8751492009080959。推測原因是此方法係直接將離群值剔除，而採「調整離群值」則仍使得資料增加一些極端的情況，因此本方法對資料之訓練會有較佳結果。
